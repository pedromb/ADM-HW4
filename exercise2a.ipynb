```python
import matplotlib.pyplot as plt
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from src.plot_ly import plot_ly
from src.graph import Graph
```

```python
reduced = False
graph = Graph(reduced)
conf_id = 14805
subgraph = graph.get_subgraph_conf(conf_id)
measures = graph.get_centralities(subgraph)
```

```python
fig, axis = plt.subplots(1, 3, figsize=(18,6))
measures_names = ["Degree", "Closeness", 'Betweenness']
for index, value in enumerate(measures):
    axis[index].hist(list(measures[index].values()), facecolor='blue', alpha=0.75)
    axis[index].set_xlabel("{}".format(measures_names[index]))
    axis[index].set_ylabel("Frequency")
    axis[index].set_title("{} Histogram".format(measures_names[index]))
```
